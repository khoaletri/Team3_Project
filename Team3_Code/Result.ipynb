```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('VOF.L.csv', index_col= 'Date', parse_dates=True, usecols=[0,4])
```

```python
df20=pd.read_csv('Result\Result20.csv', index_col='Date', parse_dates=True)
df20
```

```python
plt.figure(figsize=(12,6.5))
plt.plot(df.index, df.Close, label='Observations')
plt.plot(df20.index, df20.LNReg, label='Linear Regression')
plt.plot(df20.index, df20.NLNReg, label='Nonlinear Regression')
plt.plot(df20.index, df20.LSTM, label='LSTM')
plt.plot(df20.index, df20.Arima, label='ARIMA')
plt.plot(df20.index, df20.Prophet, label='Prophet')
plt.ylabel('Close')
plt.xlabel('Date')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.plot(df20.Close, label='Observations')
plt.plot(df20.LNReg, label='Linear Regression')
plt.plot(df20.NLNReg, label='Nonlinear Regression')
plt.plot(df20.LSTM, label='LSTM')
plt.plot(df20.Arima, label='ARIMA')
plt.plot(df20.Prophet, label='Prophet')
plt.ylabel('Close')
plt.xlabel('Date')
plt.legend()
plt.show()
```
